# Задание 1





Для оценки статистической значимости различий между конверсиями в двух группах выберем критерий статистики Хи-квадрат:

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
observed = np.array([[164, 228], [15386, 15322]])

statistic, p_value, _, _ = chi2_contingency(observed, correction=False)

alpha = 0.05

if p_value < alpha:
    result = "Есть статистически значимые различия"
else:
    result = "Нет статистически значимых различий"

print(result)
print("Статистика хи-квадрат:", statistic)
print("Значение p:", p_value)
```

Ответ: изменения внесенные в заказную форму в магазине Утконос дали статистически значимый эффект на уровень конверсии в покупку. Рекомендуется использовать новую форму заказа.
